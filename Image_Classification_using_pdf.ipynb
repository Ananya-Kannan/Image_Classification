```python
import os
import fitz  # PyMuPDF

def convert_pdf_to_images(pdf_path, output_folder):
    doc = fitz.open(pdf_path)
    for page_number in range(doc.page_count):
        page = doc.load_page(page_number)
        image = page.get_pixmap()
        image_path = os.path.join(output_folder, f"page_{page_number + 1}.png")
        image.save(image_path)
    doc.close()

if __name__ == "__main__":
    pdf_file_path = "blog on transformer based models.pdf"  # Replace with the path to your PDF file
    output_folder_path = "VIT"  # Replace with the desired output folder path

    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)

    convert_pdf_to_images(pdf_file_path, output_folder_path)
```

```python
image_array = img.open('page_1.png')
```

```python
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
inputs = feature_extractor(images = image_array, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
```

```python
predicted_class_idx = logits.argmax(-1).item()
```

```python
print(predicted_class_idx)
```

```python
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
